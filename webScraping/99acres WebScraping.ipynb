```python
from openpyxl import Workbook
from selenium.webdriver.common.by import By
import selenium
import xlsxwriter
from selenium import webdriver as wb
from webdriver_manager.chrome import ChromeDriverManager
webD=wb.Chrome('F:\chromedriver_win32\\chromedriver.exe') # here give the path of downloaded zip file
webD.get("https://www.99acres.com//")
from selenium import webdriver
import time
time.sleep(0)
 

 
element_list = []
data_set=[]
 
for page in range(160,220,1):
   # Copy the link given in href (Next button)
    page_url = "https://www.99acres.com/property-in-jaipur-ffid-page-"+ str(page)
    webD.get(page_url)
    price= webD.find_elements(By.XPATH,"//td[contains(@id,'srp_tuple_price')]") # Give ID only (srp_tuple_price)
    sqft= webD.find_elements(By.XPATH,"//td[contains(@id,'srp_tuple_primary_area')]")
    rooms= webD.find_elements(By.XPATH,"//td[contains(@id,'srp_tuple_bedroom')]")
    des=webD.find_elements(By.XPATH,"//a[contains(@id,'srp_tuple_property_title')]")
    time.sleep(1)
    
 
    for i in range(len(price)):
        element_list.append([price[i].text, sqft[i].text,rooms[i].text,des[i].text])
        data_set.append({"Price":price[i].text, "SQFT":sqft[i].text,"rooms":rooms[i].text,'Des':des[i].text})

#with xlsxwriter.Workbook('result123.xlsx') as workbook:
#    worksheet = workbook.add_worksheet()
 
#    for row_num, data in enumerate(element_list):
#        worksheet.write_row(row_num, 0, data)
       
   
```
